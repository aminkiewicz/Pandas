## AirBnB Project

The aim of this project is to clean data and then analyse it using Pandas functions. In this project, the "Seattle Airbnb Open Data" dataset from kaggle was used.

```python
import pandas as pd

pd.options.display.max_columns = None

df = pd.read_excel(r"C:\Users\adria\OneDrive\Desktop\Projects\AirBnB Project.xlsx")

df
```

**1. Finding out missing values in each column and storing them in the *missing_values* variable:**

```python
missing_values = df.isnull().sum()

missing_values
```

**2. Finding the total missing values in the whole dataset:**

```python
missing_values.sum()
```

**3. Dropping rows with missing values and storing the result in the *df_rows_dropped* variable:**

```python
df_rows_dropped = df.dropna(axis=0)

df_rows_dropped
```

**4. Dropping the column *square_feet* as it has the biggest number of values missing:**

```python
df.drop(columns='square_feet', inplace=True)

df
```

**5. Dropping columns with more than 100 missing values and storing the result in the *df_cols_dropped* variable:**

```python
df_cols_dropped = df.loc[:, missing_values <= 100]

df_cols_dropped.isnull().sum()
```

**6. Dropping all columns containing *review* in their name and storing the result in the *df_rev_cols_dropped* variable:**

```python
df = df.drop(df.filter(regex='review').columns, axis=1)

df
```

**7. Filling missing values in the *bathrooms* column with 1:**

```python
df['bathrooms'].fillna(1, inplace=True)

print(df['bathrooms'].to_string())
```

**8. Filling missing values in the *cleaning_fee* column with the min value of the column:**

```python
df['cleaning_fee'].fillna(df['cleaning_fee'].min(), inplace=True)

print(df['cleaning_fee'].to_string())
```

**9. Filling missing values in the *weekly_price* column with $50 substracted from the value in the *price* column multiplied by 7:**

```python
df['weekly_price'].fillna((df['price'] * 7) - 50.0, inplace=True)

print(df['weekly_price'].to_string())
```

**10. Filling missing values in the *monthly_price* column with the mean value of the column:**

```python
df['monthly_price'].fillna(df['monthly_price'].mean(), inplace=True)

print(df['monthly_price'].to_string())
```

**11. Filling missing values in the *security_deposit* column with the mode value of the column:**

```python
df['security_deposit'].fillna(df['security_deposit'].mode()[0], inplace=True)

print(df['security_deposit'].to_string())
```

**12. Rounding decimals in the *monthly_price* column:**

```python
df['monthly_price'] = df['monthly_price'].round(1)

print(df['monthly_price'].to_string())
```

**13. Converting strings in the *state* column to upper case:**

```python
df['state'] = df['state'].str.upper()

df['state']
```

**14. Setting index to the *id* column:**

```python
df = df.set_index('id')

df
```

**15. Checking if there are different cities than *Seattle* in the dataset:**

```python
cities = ['Seattle']

new_df = df[~df['city'].isin(cities)]

new_df.filter(items=['city'])
```

**16. Changing the city name for *id 4581256* to *Seattle*:**

```python
df.loc[4581256, 'city'] = 'Seattle'

df.filter(like = '4581256', axis = 0)
```

**17. Sorting the dataset by *id* from the lowest to the highest:**

```python
df = df.sort_index()

df
```

**18. Sorting the dataset by *price* ascending and *maximum_nights* descending:**

```python
df = df.sort_values(by=['price','maximum_nights'], ascending=[True,False])

df
```

**19. Finding the average *weekly price* for all the listings:**

```python
df['weekly_price'].mean()
```

**20. Finding the listing with the highest *price per one night*:**

```python
df['price'].max()
```

**21. Finding the listing where the *property type* is either a *boat* or a *treehouse*:**

```python
df[(df['property_type']=='Boat') | (df['property_type']=='Treehouse')]
```

**22. Finding the listing for five people, with three bedrooms, and the price is lower than or equal to *$200* per night:**

```python
df[(df['guests_included']==5) & (df['bedrooms']==3) & (df['price']<=200)]
```

**23. Finding the number of listings each host have:**

```python
df.groupby('host_id')['listing_url'].count()
```

**24. Finding hosts with more than 5 listings:**

```python
df_multiple_listings = df.groupby('host_id').filter(lambda x: x['host_id'].count() > 5)

print(df_multiple_listings.groupby('host_id')['listing_url'].count().to_string())
```

**25. Finding hosts from *New York*:**

```python
ny_hosts = df.dropna(subset=['host_location'])

ny_hosts = ny_hosts[ny_hosts['host_location'].str.contains('New York')]

ny_hosts.filter(items=['host_id', 'host_name', 'host_location'])
```
